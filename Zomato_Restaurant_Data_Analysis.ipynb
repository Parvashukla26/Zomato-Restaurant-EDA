```python
df = df.dropna(subset=['address', 'zipcode','cuisines','timings','opentable_support'])
df
```

```python
df = df.fillna("")
df
```

```python
df
```

```python
df=df.drop(columns ='url')
df=df.drop(columns ='country_id')
df=df.drop(columns = 'city_id')
df=df.drop(columns ='currency')
df=df.drop(columns ='opentable_support')
df=df.drop(columns ='delivery')
df=df.drop(columns ='takeaway')
df=df.drop(columns ='timings')
df
```

```python
restaurants = df.drop_duplicates(subset='name')
top_10_Restaurant_rating = restaurants.sort_values( by='aggregate_rating',ascending=False)[['name', 'city','aggregate_rating']].head(10)
top_10_Restaurant_rating.reset_index(drop=True)
```

```python
plt.figure(figsize=(12,6))
plt.hlines(y=top_10_Restaurant_rating['name'], xmin=0, xmax=top_10_Restaurant_rating['aggregate_rating'], color='skyblue')
plt.plot(top_10_Restaurant_rating['aggregate_rating'], top_10_Restaurant_rating['name'], "o", color='green')

for index, value in enumerate(top_10_Restaurant_rating['aggregate_rating']):
    plt.text(value + 0.05, index, f"{value:.1f}", va='center')

plt.title("Top 10 Restaurants in India by Rating", fontsize=16, fontweight='bold')
plt.xlabel("Aggregate Rating")
plt.ylabel("Restaurant Name")
plt.xlim(0, 5.5)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df['average_cost_for_two'] = pd.to_numeric(df['average_cost_for_two'], errors='coerce')
df = df[df['average_cost_for_two'] > 0].dropna(subset=['average_cost_for_two'])
sorted_df = df.sort_values(by='average_cost_for_two')
unique_price_restaurants = sorted_df.drop_duplicates(subset='average_cost_for_two')
unique_price_restaurants = unique_price_restaurants.drop_duplicates(subset='name')
cheapest_10 = unique_price_restaurants[['name', 'city', 'average_cost_for_two']].head(10).copy()
cheapest_10['Restaurant'] = cheapest_10['name'] + ' (' + cheapest_10['city'] + ')'
cheapest_10.reset_index(drop=True)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=cheapest_10,x='Restaurant',y='average_cost_for_two',hue='Restaurant',palette='coolwarm',legend = False)
plt.title('Top 10 Cheapest Restaurants in India', fontsize=16)
plt.ylabel('Price for Two (INR)')
plt.xlabel('Restaurant (City)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
ahmedabad_df = df[df['city'].str.lower() == 'ahmedabad']
top_ahmedabad_votes = ahmedabad_df.sort_values(by='votes', ascending=False).drop_duplicates(subset='name').head(10)
top_ahmedabad_votes['Name with Votes'] = top_ahmedabad_votes['name'] + " (" + top_ahmedabad_votes['votes'].astype(str) + " votes)"
top_ahmedabad_votes
```

```python
top_votes = df[df['city'].str.lower() == 'ahmedabad'].sort_values(by='votes', ascending=False).drop_duplicates(subset='name').head(10)
```
